## 1. $\textit{3F Duality}$ (3Fx: $x \notin 1,13,22,34$)

## (3F14)
### Define $T:\P(\R) \rightarrow \P(\R)$ by $(Tp)(x)=x^2p(x)+p''(x)$ for $x\in\R$. 

### (a) Suppose $\varphi \in \P(\R)'$ is defined by $\varphi(p) = p'(4)$.  Describe the linear functional $T'(\varphi)$ on $\P(\R)$.

### (b) Suppose $\varphi \in \P(\R)'$ is defined by $\varphi(p) = \int_0^1p(x)dx$. Evaluate $(T'(\varphi))(x^3)$.

#####  Recall:
Example 3.100.

#### Solution:
(a) Suppose $\varphi \in \P(\R)'$ is defined by $\varphi(p) = p'(4)$.  Describe the linear functional $T'(\varphi)$ on $\P(\R)$.

Define $$(Tp)(x)=x^2p(x)+p''(x)$$ for $x\in\R,$ then $$(T'(\varphi))(p)=(\varphi \circ T)(p)=\varphi(Tp)=\varphi p'$$
$$ \begin{aligned}\implies \varphi p'
&=\varphi(x^2 p(x)+p''(x))\\
&=(x^2 p(x)+p''(x))'(4)\\
&= (2xp(x)+x^2p'(x)+p'''(x))(4) \text{ (by product rule)}\\
&= 8p(4)+16p'(4)+p'''(4). \end{aligned}$$
$\square$

#### Solution:
(b) Suppose $\varphi \in \P(\R)'$ is defined by $\varphi(p) = \int_0^1p(x)dx$. Evaluate $(T'(\varphi))(x^3)$.
 
Define $$(Tp)=x^2p+p'' \leftarrow \text{(drop the x's as the context is clear)}$$ for $x\in\R,$ then 
$$(T'(\varphi))(p)=(\varphi \circ T)(p)=\varphi(Tp) $$
$$ \begin{aligned}\implies \varphi(Tp)
&=\varphi(x^2 p+p'')\\
&= \varphi(x^5 + 6x) \\ 
&= \int_0^1 (x^5 + 6x) \,dx \\ 
&= \left( \frac{x^6}{6} + 3x^2 \right) \Bigg|_{0}^{1}=\frac{1}{6}+3  \end{aligned} $$

$\square$

## 2. $\textit{5C Eigenspaces and Diagonal Matrices}$ (5Cx: $x \notin 3,8,10,16$)


## (5C5)
### Suppose $V$ is a finite-dimensional complex vector space and $T\in\L(V)$.  Prove that $T$ is diagonalizable if and only if $$V=\null(T-\lambda I) \oplus \range(T-\lambda I)$$ for every $\lambda \in \C$.

##### Recall: 5.41 Conditions equivalent to to diagonalizablity
Suppose $V$ is finite-dimensional and $T\in\L(V)$. Let $\v \lambda m$ denote the distinct eigen values of $T$.  Then the following are equivalent:

a) $T$ is diagonalizable.

b) $V$ has a basis consisting of eigenvectors of $T$.

c) There exist 1-dimensional subspaces $\v Un$ of $V$, each invarient under $T$ such that $$V=U_1\oplus\cdots\oplus U_n.$$

d) $V=E(\lambda_1)\oplus\cdots\oplus E(\lambda_m)$.

#### Solution:
$\Rightarrow$

Let $\T$ defined by $(T-\lambda I)$ and suppose $\T\in\L(V)$ is diagonalizable.

An operator $\T\in\L(V)$ has a diagonal matrix
$$\begin{pmatrix} \lambda_1 & & 0 \\ &\ddots& \\ 0& & \lambda_n\end{pmatrix}$$
with respect to a basis $\v vn$ of $V$ if and only if $\T v_j=\lambda_j v_j$ for each $j$.

Thus $\lambda_j=0$ is the basis of $\null \T$ and all $v_j$ such that $\lambda_j \neq 0$ are a basis of $\range \T$. 

[It's worth noting here that this is a basis with respect to $\T$ rather than $T$.]

Hence if $T$ is diagonalizable then $\T=T-\lambda I$ is also diagonalizable for $\lambda\in\C$.  

$\square$

$\Leftarrow$

Consider a contrary case where $\T$ is invertible thus not diagonalizable.  Recall
$$V=\null(\T) \oplus \range(\T).$$
For this to be true $\lambda \notin \span \T$, in other words $\v \lambda m$ are not distinct eigenvalues of $\T$.

Hence if $T$ is diagonalizable then $\lambda \neq \lambda_j$.

$\square\square$

## 3. $\textit{6B Orthonormal Basis}$ (6Bx: $x\notin 2,5,15$)


## (6B17)

### For $u\in V$, let $\Phi u$ denote the linear functional on $V$ defined by $$(\Phi u)(v)=\ip{v,u}$$ for $v\in V$.

### (a) Show that is $\F=\R$, the $\Phi$ is a linear map from $V$ to $V'$. (Recall from Section 3.F that $V'=\L(V,\F)$ and that $V'$ is called the dual space of $V$.)

### (b) Show that if $\F=\C$ and $V\neq\0$, then $\Phi$ is not a linear map.

### (c) Show that $\Phi$ is injective.

### (d) Suppose $\F=\R$ and $V$ is finite-dimensional.  Use parts (a) and (c) and a dimension-counting argument (but without using 6.42) to show that $\Phi$ is an isomorphism from $V$ onto $V'$.

[Part (d) gives and alternative proof of the Riesz Representation Theorem (6.42) when $\F=\R$. Part (d) also gives a natural isomorphism (meaning that it does not depend on a choise of basis) from a finite-dimensional real inner product space onto its dual space.]

##### Recall: Linear Map
A linear map from $V$ to $W$ is a function $T:V \rightarrow W$ with the following properties:

Additivity: $T(u+v)=Tu + Tv$ for all $u,v\in V$.

Homogeneity: $T(\lambda v)=\lambda(Tv)$ for all $\lambda\in\F$ and all $v\in V$.

#### Solution: 
(a) Show that is $\F=\R$, the $\Phi$ is a linear map from $V$ to $V'$. (Recall from Section 3.F that $V'=\L(V,\F)$ and that $V'$ is called the dual space of $V$.)

$\Rightarrow$
Additiviy:

Suppose $u_1,u_2 \in V$ and let $\Phi u$ denote the linear functional on $V$ defined by $$(\Phi u)(v)=\ip{v,u}$$ for $u,v\in V$.

Then, 
$$ \begin{aligned}
(\Phi(u_1 + u_2))(v) 
&= \langle v, u_1 + u_2 \rangle\\
&= \langle v, u_1 \rangle + \langle v, u_2 \rangle\\
&= (\Phi u_1)(v) + (\Phi u_2)(v). \end{aligned}
$$

$\Leftarrow$
Homogeneity:

Suppose $u,v \in V$, $c \in \mathbb{R}$ and let $\Phi u$ denote the linear functional on $V$ defined by $$(\Phi u)(v)=\ip{v,u}$$ for $u,v\in V$.

Then,
$$ \begin{aligned}
(\Phi(cu))(v) &=\langle v, cu \rangle\\
&= c\langle v, u\rangle \\
&= c(\Phi u)(v).
\end{aligned}$$

Thus $\Phi$ is a linear map from $V$ to $V'$

$\square$

#### Solution:
(b) Show that if $\F=\C$ and $V\neq\0$, then $\Phi$ is not a linear map.

Consider $\F=\C$, then
$$(\Phi(cu))(v) = \bar{c}(\Phi u)(v)$$

$$\implies c = \bar{c}.$$

However this can only be true if $c=0$ or $c\in\R$, thus $\Phi$ is not a linear map.

$\square$

##### Recall: Injective
A function $T:V\rightarrow W$ is injective is $Tu=Tv \implies u=v$.

#### Solution:
(c) Show that $\Phi$ is injective.

Let $\Phi u$ denote the linear functional on $V$ defined by $$(\Phi u)(v)=\ip{v,u}$$ for $u,v\in V$.

Suppose there are $u_1$ and $u_2$ in $V$ such that $\Phi u_1 = \Phi u_2$.
Then
$$0 = (\Phi u_1-\Phi u_2)(v) = (\Phi(u_1-u_2))(v) = \langle v, u_1-u_2 \rangle,$$
thus $\Phi$ is injective.

$\square$

##### See also:
6.42 Riesz Representation Theorem and 6.30 Linear Combination of Orthonormal Basis.

##### Recall: Fundamental Theorem of Linear Maps
Suppose $V$ is finite-dimensional and $T\in\L(V,W)$.  Then $\range T$ is finite-dimensional and 
$$\dim V = \dim \null T + \dim \range T.$$

##### Recall: $T$ surjective is equivalent to $T'$ injective
Suppose $V$ and $W$ are finite-dimensional and $T\in\L(V,W)$.  Then $T$ is surjective if and only if $T'$ is injective.

##### Recall: $T$ injective is equivalent to $T'$ surjective  
Suppose $V$ and $W$ are finite-dimensional and $T\in\L(V,W)$.  Then $T$ is injective if and only if $T'$ is sujective.

##### Recall: Dimension shows whether vector spaces are isomorphic
Two finite-dimensional vector spaces over $\F$ are isomorphic if and only if they have the same dimension.

#### Solution:
(d) Suppose $\F=\R$ and $V$ is finite-dimensional.  Use parts (a) and (c) and a dimension-counting argument (but without using 6.42) to show that $\Phi$ is an isomorphism from $V$ onto $V'$.

By (a), we know $\Phi$ is a linear map from $V$ to $V'$ and by (c), we showed that $\dim \null \Phi = 0$.

Then, by the Fundamental Theorem of Linear Maps,
$$ \begin{aligned}
\dim V &= \dim \null \Phi + \dim \range \Phi\\
&=\dim \range \Phi\\
&=\dim V'.
\end{aligned}$$

Thus, $\Phi$ isomorphism from $V$ to $V'$.

$\square$

##### Recall: Gram-Schmidt Procedure
Suppose $\v vm$ is linearly independent list of vectors in $V$. Let $e_e=v_1/\|v_1\|$.  For $j=2,...m$, define $e_j$ inductively by
$$e_j=\frac{v_j - \ip{v_j,e_1}e_1 - ... - \ip{v_j,e_{j-1}}e_{j-1}}{\| v_j - \ip{v_j,e_1}e_1 - ... - \ip{v_j,e_{j-1}}e_{j-1} \|}.$$
Then $\v em$ is an orthonormal list of vector in $V$ such that
$$\span(\v vj)=\span(\v ej)$$ for $j=1,...,m$.

````python
# find our basis
n = 5
x = sym.symbols('x', real=True)
v = [x**i for i in range(n+1)]
lim0 = -sym.pi
lim1 = sym.pi

# inner-product lambda function
ip = lambda f,g: sym.integrate(f*g, (x,lim0,lim1))*g

# orthonomal basis procedure (GramSchmidt)
def basis(v):
    e = []
    for i, vi in enumerate(v):
        ei_nom = (vi - sum([ip(vi,ej) for ej in e]))
        #ei_denom = sym.sqrt(ip(ei_nom,ei_nom)) # very slow
        ei_denom = sym.sqrt(sym.integrate(ei_nom**2, (x,lim0,lim1))) # much better
        ei = ei_nom/ei_denom
        e.append(ei)
    return e

# show our basis
e = basis(v)
display(e) # Our polynominal basis...
````

## 5. $\textit{7A Self-Adjoint and Normal Operators}$ (7Ax: $\forall x$)
## (7A20)
### Suppose $T\L(V,W)$ and $\F=\R$.  Let $\Phi_V$ be the isomorphisim from $V$ onto the dual space $V'$ given by Exercise 17 in Section 6.B, and let $\Phi_W$ be the corresponding isomorphism from $W$ onto $W'$.  Show that if $\Phi_V$ and $\Phi_W$ are used to identify $V$ and $W$ with $V'$ and $W'$, then $T^*$ is identified with the dual map $T'$.  More precisely, show that $\Phi_V \circ T^* = T' \circ \Phi_W$.

##### Recall: Riesz Representation Theorem
Suppose $V$ is finite-dimensional and $\varphi$ is a linear functional on $V$.  Then there is a unique vector $u \in V$ such that $$\varphi=\ip{v,u}$$ for every $v\in V$.

#### Solution:
Let $v \in V$ and $w \in W$, show that $$\Phi_V \circ T^* = T' \circ \Phi_W.$$

$\Leftarrow$ LHS

$$\begin{aligned}
((\Phi_V \circ T^*)(w))(v) &= (\Phi_V(T^*w))(v)\\\\
&= \langle v, T^*v \rangle\\\\
&= \langle Tv, w \rangle.
\end{aligned}$$

$\Rightarrow$ RHS

$$\begin{aligned}
((T' \circ \Phi_W)(w))(v) &= (T' \circ \Phi_W(w))(v)\\\\
&= (\Phi_W(w) \circ T)(v)\\\\
&= (\Phi_W(w))(Tv)\\\\
&= \langle Tv, w \rangle.
\end{aligned}$$

Thus 
$$\begin{aligned}
((\Phi_V \circ T^*)(w))(v) &= (T' \circ \Phi_W)(w))(v) \\
\implies ((\Phi_V \circ T^*) &= (T' \circ \Phi_W),
\end{aligned}$$
as desired.

$\square$

##### Recall: Definition of Normal
 - an operator on an inner production is called normal if it commutes with its adjoint
 - in other words, $T\in\L(V)$ is normal if $$TT^*=T^*T.$$

#### Solution:

Let $T=\left[\begin{matrix}1 & 1 & 0\\0 & 1 & 1\\1 & 0 & x\end{matrix}\right]$. By direct computation,
$$TT^*=T^*T.$$
$$\left[\begin{matrix}2 & 1 & 1\\1 & 2 & \overline{x}\\1 & x & x \overline{x} + 1\end{matrix}\right]=\left[\begin{matrix}2 & 1 & x\\1 & 2 & 1\\\overline{x} & 1 & x \overline{x} + 1\end{matrix}\right]$$
$$\implies \overline{x}=1 \text{ and } x=1.$$
Hence the covered number is $1$.

$\square$
