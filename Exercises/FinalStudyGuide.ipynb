```python
n = 16
omega = -sym.I * sym.pi/n 
sym.exp(omega)
V = Matrix([sym.exp(omega)**i for i in range(n)])
M = sym.diag(V.tolist(), unpack=True)
#V,M
```

```python
# Build a Matrix for our basis using permutations M^k
B = Matrix([[(M**k)*sym.ones(n,1) for k in range(n)]])  # not so efficent
#B, B.rref() #notice this is a polynomial basis
B
```

```python
#plt.quiver(X, Y, U, V, color='b', units='xy', scale=1)
#fig = plt.figure(figsize = (15, 15))
for x in range(n):
    for y in range(n):
        z = complex(B[x,y])
        #display(z.real)
        plt.quiver(x, y, z.real,z.imag, color='b', angles='xy', scale_units='xy', scale=1)    
plt.grid()

# x-lim and y-lim
plt.xlim(0-1, n+1)
plt.ylim(0-1, n+1)
```

```python
import numpy as np

N = 30
e = np.exp(1j*2*np.pi/N)

s = 0
points = [[0, 0]]
arrows = []
for k in range(N):
    s += e**k
    point = [s.real, s.imag]
    arrow = [(e**k).real, (e**k).imag]
    points.append(point)
    arrows.append(arrow)
    
points = np.array(points[:-1])
arrows = np.array(arrows)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.gca()

ax.quiver(points[:, 0], points[:, 1], 
          arrows[:, 0], arrows[:, 1], 
          angles='xy', scale_units='xy', scale=1)

ax.text(-0.8, 4.5, r'$\sum_{k=0}^{N-1} e^{2\pi i k / N}$', 
        bbox=dict(boxstyle='square,pad=1.5', fc='lightgrey'))

ax.set_aspect('equal')
ax.grid()
ax.set_xlabel('real part')
ax.set_ylabel('imaginary part')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

t = np.linspace(0, 20, 30)
x = np.linspace(-2*np.pi, 2*np.pi, 31)
T, X = np.meshgrid(t, x)

DT = np.ones_like(X)
DX = f(X)

fig = plt.figure()
ax = fig.gca()

ax.quiver(T, X, DT, DX)

ax.set_xlabel('t')
ax.set_ylabel('x')
```

```python
x =  y = np.linspace(-5, 5, 40)
q=1

X, Y = np.meshgrid(x, y)

a = np.array([1.5, 0, 0])

rma = np.array([X - a[0], Y + a[1]])
rpa = np.array([X + a[0], Y + a[1]])

E_dip = q * rma / np.sqrt(rma[0, ...]**2 + rma[0, ...]**2)**3
E_dip -= q * rpa / np.sqrt(rpa[0, ...]**2 + rpa[0, ...]**2)**3

# Normalize the arrows because we have singularities where the
# charges are located:

norms = np.sqrt(E_dip[0, ...]**2 + E_dip[1, ...]**2)
E_dip /= norms 

plt.quiver(X, Y, E_dip[0, ...], E_dip[1, ...], minshaft=2)
```
