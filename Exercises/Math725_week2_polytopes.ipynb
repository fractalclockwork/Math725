```python
def m_plot(Z):  # assumes matrix
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    z=np.array((Z.row(0)).tolist()).astype(np.complex64) # Aww Yeah!!! This is fast.
    #ax.scatter3D(z.real, z.imag, 1*np.ones(len(z)) )
    #ax.scatter3D(z.real, z.imag, -1*np.ones(len(z)) )
    ax.scatter3D(z.real, z.imag, np.zeros(len(z)) )
```

```python
# made a simple prototype to visualize, how many licks does it take to get to the centre...
n=4 # number of points on each face, ie 4 for cube
zeta = sym.exp(sym.I*2*sym.pi/n)
shape = Matrix([zeta**i for i in range(n)]).T 
m_plot(shape) 
```

$S^2=\{(x,y,x)\in \Bbb{R}^3| x^2+y^2+z^2=1\}$
