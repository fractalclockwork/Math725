### Def: Span
The set of all linear combinations of a list of vectors $v_1,...,v_m$ in $V$ is call the $\textbf{span}$ of $v_1,...,v_m$, denoted span($v_1,...,v_m$).  In other words,

span($v_1,...,v_m$) $= \{a_1v_1+...+a_mv_m: a1,...,a_m \in \Bbb{F}\}$

The span of the empty list $()$ is defined to be $\{0\}$.

$v_1,...,v_m$  is a basis of $V$ iff,

$\forall v\in V, v=a_1v_1+...+a_nv_n$, where $a_1,...,a_m \in \Bbb{F}$ and the choice of $a_1,...,a_m$ is unique.

 * Existance is guaranteed by spanning property of $v_1,...,v_m$.
