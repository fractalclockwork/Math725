### Theorem:  

If $U\subset V$ is finite dimensional, then $U$ is also finite dimensional.

```python
import schemdraw # requires schemdraw==0.14, 0.15 isn't compatible with current matplotlib
from schemdraw import flow

display(Latex('If $U\subset V$ is finite dimensional, then $U$ is also finite dimensional.'))

with schemdraw.Drawing() as d:
    d += flow.Start().label('START')
    d += flow.Arrow().down(d.unit/3)
    
    d += (d0 := flow.Decision(w=3, h=2, E='YES',S='$v_1\in u$').label('u?={0}'))
    d += flow.Line().down(d.unit/4)
    d += (d1 := flow.Decision(w=4, h=2, E='YES', S='$\exists_{v_2}\in$ span$(v_1)$').label('span $v_1 ?= u$'))
    d += flow.Line().down(d.unit/4)
    d += (d2 := flow.Decision(w=5, h=2, E='YES',S='$\exists_{v_3}\in$ span$(v_1,v_2)$').label('span $(v_1,v_2) ?= u$'))
    d += flow.Line().down(d.unit/4)
    d += (d3 := flow.Decision(w=3, h=2).label('...'))

    
    d += flow.Arrow().right(d.unit/2).at(d0.E)
    d += flow.Box(w=2, h=1.25).anchor('W').label('DONE')
    d += flow.Arrow().right(d.unit/2).at(d1.E)
    d += flow.Box(w=2, h=1.25).anchor('W').label('DONE')
    d += flow.Arrow().right(d.unit/2).at(d2.E)
    d += flow.Box(w=2, h=1.25).anchor('W').label('DONE')
display(Latex('$v_1,...,v_k$ is linearly independent, thus by L2.23 the process terminates.'))
```

### Def: Span
The set of all linear combinations of a list of vectors $v_1,...,v_m$ in $V$ is call the $\textbf{span}$ of $v_1,...,v_m$, denoted span($v_1,...,v_m$).  In other words,

span($v_1,...,v_m$) $= \{a_1v_1+...+a_mv_m: a1,...,a_m \in \Bbb{F}\}$

The span of the empty list $()$ is defined to be $\{0\}$.
